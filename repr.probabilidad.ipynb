```python
import numpy as np
import matplotlib.pyplot as plt

# Carga los datos
datos = np.loadtxt("probabilidades_pd.txt")

# Primer columna: tiempo (n), segunda: PD
tiempo = datos[:, 0]
PD = datos[:, 1]

plt.figure(figsize=(8,4))
plt.plot(tiempo, PD, marker='o', linestyle='', color='blue')
plt.xlabel("Tiempo (n)")
plt.ylabel("Probabilidad PD")
plt.title("Probabilidad de transmisión en función del tiempo")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

datos = np.loadtxt("j_esperado.txt")
tiempo = datos[:,0]
x_esp = datos[:,1]

plt.plot(tiempo, x_esp)
plt.xlabel("Tiempo")
plt.ylabel("<j>(t)")
plt.title("Valor esperado de la posición en función del tiempo")
plt.grid()
plt.show()
```
